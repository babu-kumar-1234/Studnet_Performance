## LinearRegression b/w Previous Scores vs Performance Index

```python
df.corr()
```

```python
sns.heatmap(df.corr(),cmap = 'YlGnBu', annot = True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df['Previous Scores']
y = df['Performance Index']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.8,random_state=10)
```

```python
x_train.head()
```

```python
x_train.sum()
```

```python
x_test.sum()
```

```python
y_train.sum()
```

```python
y_test.sum()
```

```python
x_train = x_train.values.reshape(-1,1)
x_test = x_test.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
print(lm.intercept_) # for 'c' value
```

```python
print(lm.coef_) # for 'm' value
```

```python
# visualize regression line for best line
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,-15.14+1.01*x_train,'r')
```

```python
# error analysis for (traing data)
```

```python
y_train_pred = lm.predict(x_train)
res_train = (y_train - y_train_pred)
```

```python
sns.distplot(res_train,bins=15)
plt.title('error terms')
plt.xlabel('residual error')
plt.ylabel('freq')
```

```python
# error for (testing data)
```

```python
y_test_pred = lm.predict(x_test)
res_test = (y_test - y_test_pred)
```

```python
sns.distplot(res_test,bins=10)
plt.title('test error terms')
plt.xlabel('residual error')
plt.ylabel('freq')
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

### r2_score

```python
print(r2_score(y_train,y_train_pred))
```

```python
print(r2_score(y_test,y_test_pred))
```

- Training R2 Score: 0.8374
- Testing R2 Score: 0.8379

Interpretation:
- Good model performance: Both R2 scores are around 0.84, indicating that about 84% of the variance in the data is explained by the model.
- Consistency: The similarity between training and testing R2 scores suggests that the model generalizes well and isn't overfitting.

## Conclusion 

* Based on the R2 scores provided for each feature and target pair, the best feature for linear regression with the Performance Index as the target variable is:

> Previous Scores
* The R2 scores for the Linear Regression between Previous Scores and Performance Index are:

- Training R2 Score: 0.8374
- Testing R2 Score: 0.8379

> These high R2 scores indicate that the model explains a significant portion of the variance in the Performance Index, suggesting a strong linear relationship between Previous Scores and Performance Index.

* Why Previous Scores is the best feature:
1. High R2 scores: The R2 scores are significantly higher than those of other features, indicating a strong relationship between Previous Scores and Performance Index.
2. Consistency: Both training and testing R2 scores are high and consistent, suggesting that the model generalizes well.

* Comparison with other features:
- Hours Studied: While Hours Studied has a moderate R2 score (around 0.14), it's still much lower than Previous Scores.
- Other features: The R2 scores for Sample Question Papers Practiced, Sleep Hours, and Extracurricular Activities are very low (close to 0), indicating weak relationships with Performance Index.

* Based on these results, Previous Scores is the most suitable feature for linear regression with Performance Index as the target variable.
